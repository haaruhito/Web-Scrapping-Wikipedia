<mark> From the html file, we can see there is a word 'table' from where the table starts. 'tr' means the whole row while 'td' means a cell of the row. The table has a class 'wikitable', so lets access it. 

```python
table = soup.find('table',attrs={'class':'wikitable'})
```

```python
table
```

<mark> Above 'tr' represents the rows and 'th' in the first row is each cells in the row which is the name of the columns. In the second row ('tr'), there is 'td' which is also each cell.


<mark> First lets access all the rows from the table above and from the first row, we will access the name of the columns

```python
rows = table.find_all('tr')
```

```python
#rows
```

```python
 #rows[0].text
```

<mark> So, lets access name of each columns from row 0. We need to loop through to get each cell of the columns. 

```python
columns = [i.text.replace('\n','') for i in rows[0].find_all('th')]
```

```python
columns
```

```python
#rows
```

# Extracting images

```python
images = table.find_all('a',attrs={'class':'image'})
```

```python
images
```

<mark> From the above output, lets find the link where there is '.jpg' at the end. We can see that the link is in 'src' which we can access using the img['src']. We will add "https:" in front of the link to complete the link. 

```python
image_list = []
for i in images:
    #print(i.img['src'])
    a = 'https:' + i.img['src']
    image_list.append(a)
    
```

```python
image_list
```

<mark> Below I will extract all text from the rows and replace next line ('\n') with ','

```python
# Extracting Death Date from the 'Death' column of the dataframe above
df['Death Day'] = df.Death.str.extract(r'((?=)\d+\ +\w+\ \d+)')
```

```python
#Extracting Place of Birth from 'Death' column of the dataframe above
df['Birth Place'] = df.Death.str.replace(r'((?=)\d+\ +\w+\ \d+)','')
```

```python
# Removing extra letters i.e. 'd.' from the birth place string.
df['Birth Place'] = df['Birth Place'].str.replace(r'([d.])','')
```

<mark> From the dataframe above, we can see that the columns 'Birth and death' should be changed as it has only Birth date. Also in the 'Death' column, we can see that the data has been somewhat messed up. The 'Death' column contains the Birth Place and Death Day together. Hence, I will use regex to extract only death date and add it to a separate column. Also I will add the Birth place into another column. Finally, we wont be needing 'Death' column as we will have two extra columns for this information. 

```python
# Now rename the 'Birth and Death' column as 'Birth Date' and lets remove the Death column
```

```python
df.rename(columns={'Birth and death':'Birth Date'},inplace=True)
```

```python
df.drop(columns=['Death'],inplace=True)
```
